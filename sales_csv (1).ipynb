```python
# Step 1: Libraries import
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Sample Sales Data banaye
data = {
    'Product': ['A', 'B', 'C', 'A', 'B', 'C'],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': [100, 150, 200, 120, 130, 170]
}

# DataFrame create karo
df = pd.DataFrame(data)

# CSV file banaye (temporary)
df.to_csv('sales_data.csv', index=False)

# CSV load karke check karo
df = pd.read_csv('sales_data.csv')
print("=== Data Loaded ===")
print(df)

# Step 3: Data understanding
print("\n=== Data Info ===")
print(df.info())

print("\n=== Summary Statistics ===")
print(df.describe())

print("\n=== Missing Values Check ===")
print(df.isnull().sum())

# Step 4: Analyze sales
# Product wise total sales
sales_by_product = df.groupby('Product')['Sales'].sum()
print("\n=== Total Sales by Product ===")
print(sales_by_product)

# Region wise total sales
sales_by_region = df.groupby('Region')['Sales'].sum()
print("\n=== Total Sales by Region ===")
print(sales_by_region)

# Step 5: Visualizations
# Bar chart - Product wise sales
sales_by_product.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()

# Pie chart - Region wise sales
sales_by_region.plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales Distribution by Region')
plt.ylabel('')
plt.show()

# Step 6: Insights
print("\n=== Insights ===")
print("1. Highest selling product:", sales_by_product.idxmax())
print("2. Region with maximum sales:", sales_by_region.idxmax())
```
